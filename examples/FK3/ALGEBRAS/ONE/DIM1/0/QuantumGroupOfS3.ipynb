# El ГЎlgebra $\mathfrak{D}=\mathcal{D}(\mathcal{FK}_m\# \Bbbk\mathbb{S}_{3})$
Sea $\mathcal{FK}_m$ el ГЎlgebra presentada por generadores

$$
\{ x_{(ij)} \mid i<j \in \mathbb{I}_m \}
$$
y relaciones
$$
\begin{align*}
x_{(ij)}^2 &= 0, && i<j \in \mathbb{I}_m, \\
x_{(ij)}x_{(kl)} - x_{(kl)}x_{(ij)} &= 0, && i,j,k,l \in \mathbb{I}_m, \text{ todos distintos}, \\
x_{(jk)}x_{(ik)} - x_{(ij)}x_{(jk)} + x_{(ik)}x_{(ij)} &= 0, && i<j<k \in \mathbb{I}_m, \\
x_{(ik)}x_{(jk)} - x_{(jk)}x_{(ij)} + x_{(ij)}x_{(ik)} &= 0, && i<j<k \in \mathbb{I}_m.
\end{align*}
$$

## El ГЎlgebra $\mathcal{FK}_3$


Para $m=3$, el ГЎlgebra $\mathcal{FK}_3$ estГЎ generada por
$$
x_{(12)},\quad x_{(13)},\quad x_{(23)}
$$
y las relaciones son:
$$
\begin{aligned}
x_{(12)}^2 &= 0, \\
x_{(13)}^2 &= 0, \\
x_{(23)}^2 &= 0, \\
x_{(23)}x_{(13)} - x_{(12)}x_{(23)} + x_{(13)}x_{(12)} &= 0, \\
x_{(13)}x_{(23)} - x_{(23)}x_{(12)} + x_{(12)}x_{(13)} &= 0.
\end{aligned}
$$

## Relaciones en el ГЎlgebra $\mathcal{FK}_3$

### *Nilpotencia cuadrГЎtica*:
$$
\begin{align*}
x_{(12)}^2 = x_{(13)}^2 = x_{(23)}^2 = 0,
\end{align*}
$$
Para nosotros serГЎn:
$$
\begin{align*}
x_{12}^2 = x_{13}^2 = x_{23}^2 = 0.
\end{align*}
$$

### Relaciones triples para $i<j<k$:

- Para $1<2<3$:
$$
x^{2}_{(ij)} \;=\; x_{(ij)}x_{(ik)} \;+\; x_{(jk)}x_{(ij)} \;+\; x_{(ik)}x_{(jk)}
\;=\;
x_{(ik)}x_{(ij)} \;+\; x_{(ij)}x_{(jk)} \;+\; x_{(jk)}x_{(ik)}
\;=\; 0,
\quad \text{(dada por } B(V)).
$$

$$
\begin{aligned}
x_{(12)}x_{(13)} + x_{(23)}x_{(12)} + x_{(13)}x_{(23)} 
\quad \text{y} \quad 
x_{(13)}x_{(12)} + x_{(12)}x_{(23)} + x_{(23)}x_{(13)}.
\end{aligned}
$$

```gap
###############################################################################
# x_(ij)^2 = 0, x_(ij)x_(jk)+x_(jk)x_(ik)+x_(ik)x_(ij)=0, etc. 
# (ГЎlgebra Fomin-Kirillov n=3,
###############################################################################

# Generadores x_(ij)
gensFK3      := [ (1,2), (1,3), (2,3) ];;
generators   := gens{[2..4]};;
relationsFK3 := List( generators, s -> s^2);;
i            := 1;;
j            := 2;;
k            := 3;;
Xij          := generators[Position(gensFK3, (i,j))];;
Xik          := generators[Position(gensFK3, (i,k))];;
Xjk          := generators[Position(gensFK3, (j,k))];;
# primera relaciГіn: x_(ij)x_(jk) + x_(jk)x_(ik) + x_(ik)x_(ij) = 0
Add(relationsFK3, Xij*Xjk + Xjk*Xik + Xik*Xij );
# segunda relaciГіn: x_(jk)x_(ij) + x_(ik)x_(jk) + x_(ij)x_(ik) = 0
Add(relationsFK3, Xjk*Xij + Xik*Xjk + Xij*Xik );
```

## Relaciones en el ГЎlgebra $\mathcal{FK}_3$ dual

### *Nilpotencia cuadrГЎtica*:
$$
\begin{align*}
y_{(12)}^2 = y_{(13)}^2 = y_{(23)}^2 = 0.\\
\end{align*}
$$
Para nosotros serГЎn:
$$
\begin{align*}
y_{12}^2 = y_{13}^2 = y_{23}^2 = 0.
\end{align*}
$$

### Relaciones triples para $i<j<k$:

- Para $1<2<3$:
$$
\begin{align*}
    y_{(23)}y_{(13)} + y_{(12)}y_{(23)} + y_{(13)}y_{(12)} &= 0,&&\Rightarrow& y_3y_1 + y_0y_3 + y_1y_0 &= 0, \\
    y_{(13)}x_{(23)} + y_{(23)}y_{(12)} + y_{(12)}y_{(13)} &= 0,&&\Rightarrow& y_1y_3 + y_3y_0 + y_0y_1 &= 0.
\end{align*}
$$

```gap
###############################################################################
# y_(ij)^2 = 0, y_(ij)y_(jk)+y_(jk)y_(ik)+y_(ik)y_(ij)=0, etc. 
# (ГЎlgebra Fomin-Kirillov n=3,
###############################################################################

# Generadores x_(ij)
gensFK3     := [ (1,2), (1,3), (2,3) ];;
relsFK3dual := [];;
# Paso 1: Relaciones cuadrГЎticas x_(ij)^2 = 0
for k in [1..Length(gensFK3 )] do
    Add(relsFK3dual, [[ [yIndex[k], yIndex[k]] ], [1]]);;
od;
# Paso 2: Relaciones cГєbicas de tres Г­ndices distintos
triples := [[1,2,3]];;  # Гєnico triple distinto para n=3
for t in triples do
    i := t[1];
    j := t[2];
    k := t[3];
    ij := Position(gensFK3, (i,j));
    ik := Position(gensFK3, (i,k));
    jk := Position(gensFK3, (j,k));
    # primera relaciГіn: y_(ij)y_(jk) + y_(jk)y_(ik) + y_(ik)y_(ij) = 0
    Add(relsFK3dual,[[[yIndex[ij],yIndex[jk]],[yIndex[jk],yIndex[ik]],[yIndex[ik],yIndex[ij]]],[1,1,1]]);
    # segunda relaciГіn: y_(jk)y_(ij) + y_(ik)y_(jk) + y_(ij)y_(ik) = 0
    Add(relsFK3dual,[[[yIndex[jk],yIndex[ij]],[yIndex[ik],yIndex[jk]],[yIndex[ij],yIndex[ik]]],[1,1,1]]);
od;
```

# The simple Verma modules

Let us take the expression
$$
\Omega := -2\tau^2\delta_e + 2\tau\sigma\delta_e + 4\sigma\tau\delta_e - 2\tau\delta_e + 2\sigma\delta_e - 2\sigma\tau - 2
4$$. 
Since $\delta_e$ acts as the identity in all cases, we can formally replace
$$
\delta_e \mapsto I.
$$ 
Then
$$
\Omega = -2\tau^2 + 2\tau\sigma + 4\sigma\tau - 2\tau + 2\sigma - 2\sigma\tau - 2.
$$
Grouping:
$$
\Omega = -2\tau^2 + 2\tau\sigma + 2\sigma\tau - 2\tau + 2\sigma - 2.
$$
We will use:
$$
\sigma = (12), \quad \tau = (123).
$$
In $\mathbb{S}_3$,
$$
\tau^2 = (132),
$$
and furthermore
\begin{align*}
\tau\sigma &= (123)(12) = (13), \\
\sigma\tau &= (12) (123) = (23).
\end{align*}
Therefore,
$$
\Omega = -2(132) + 2(13) + 2(23) - 2(123) + 2(12) - 2e.
$$

## The Verma modules $\mathrm{M}(e, +)$ isn't $\mathfrak{D}$-simple.

- 1) trivial representation on $|e\rangle_{+}$

The trivial representation satisfies
$$
\varepsilon(\text{transposition}) = 1, \quad \varepsilon(\text{3-cycle}) = 1.
$$

Then:
\begin{align*}
\varepsilon(\tau) = 1, \quad \varepsilon(\tau^2) = 1, \\
\varepsilon(\sigma) = 1, \quad \varepsilon(\tau\sigma) = 1, \quad \varepsilon(\sigma\tau) = 1.
\end{align*}
We apply this term by term:
\begin{align*}
-2\tau^2 &\mapsto -2(1) = -2, \\
2\tau\sigma &\mapsto 2(1) = 2, \\
2\sigma\tau &\mapsto 2(1) = 2, \\
-2\tau &\mapsto -2(1) = -2, \\
2\sigma &\mapsto 2(1) = 2, \\
-2e &\mapsto -2.
\end{align*}
Adding up:
$$
-2 + 2 + 2 - 2 + 2 - 2 = 0.
$$
Therefore,
$$
\Omega|e\rangle_{+} = 0 |e\rangle_{-} = 0.
$$

The relation is a direct consequence of Schur's Lemma.

Given the decomposition $M \cong \bigoplus_{j=1}^n S_j^{\oplus t_j}$, applying the left-exact functor $\operatorname{Hom}_A(S_i, -)$ yields an isomorphism of vector spaces over the base field $k$:

$$\operatorname{Hom}_A(S_i, M) \cong \bigoplus_{j=1}^n \operatorname{Hom}_A(S_i, S_j)^{\oplus t_j}$$

Because the modules $\{S_1, \dots, S_n\}$ are irreducible and mutually non-isomorphic, Schur's Lemma dictates that $\operatorname{Hom}_A(S_i, S_j) = 0$ whenever $i \neq j$. Consequently, all cross-terms vanish, leaving:

$$\operatorname{Hom}_A(S_i, M) \cong \operatorname{Hom}_A(S_i, S_i)^{\oplus t_i}$$

By Schur's Lemma, $\operatorname{Hom}_A(S_i, S_i)$ is a division algebra over $k$. Taking dimensions over $k$ gives the explicit formula for the multiplicity $t_i$:

$$t_i = \frac{\dim_k \operatorname{Hom}_A(S_i, M)}{\dim_k \operatorname{Hom}_A(S_i, S_i)}$$

If the base field $k$ is algebraically closed, $\operatorname{Hom}_A(S_i, S_i) \cong k$, making its dimension $1$. In this case, the relation reduces entirely to:

$$t_i = \dim_k \operatorname{Hom}_A(S_i, M)$$
