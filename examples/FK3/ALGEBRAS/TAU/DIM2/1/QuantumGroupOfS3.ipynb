# El álgebra $\mathfrak{D}=\mathcal{D}(\mathcal{FK}_m\# \Bbbk\mathbb{S}_{3})$
Sea $\mathcal{FK}_m$ el álgebra presentada por generadores

$$
\{ x_{(ij)} \mid i<j \in \mathbb{I}_m \}
$$
y relaciones
$$
\begin{align*}
x_{(ij)}^2 &= 0, && i<j \in \mathbb{I}_m, \\
x_{(ij)}x_{(kl)} - x_{(kl)}x_{(ij)} &= 0, && i,j,k,l \in \mathbb{I}_m, \text{ todos distintos}, \\
x_{(jk)}x_{(ik)} - x_{(ij)}x_{(jk)} + x_{(ik)}x_{(ij)} &= 0, && i<j<k \in \mathbb{I}_m, \\
x_{(ik)}x_{(jk)} - x_{(jk)}x_{(ij)} + x_{(ij)}x_{(ik)} &= 0, && i<j<k \in \mathbb{I}_m.
\end{align*}
$$

## El álgebra $\mathcal{FK}_3$


Para $m=3$, el álgebra $\mathcal{FK}_3$ está generada por
$$
x_{(12)},\quad x_{(13)},\quad x_{(23)}
$$
y las relaciones son:
$$
\begin{aligned}
x_{(12)}^2 &= 0, \\
x_{(13)}^2 &= 0, \\
x_{(23)}^2 &= 0, \\
x_{(23)}x_{(13)} - x_{(12)}x_{(23)} + x_{(13)}x_{(12)} &= 0, \\
x_{(13)}x_{(23)} - x_{(23)}x_{(12)} + x_{(12)}x_{(13)} &= 0.
\end{aligned}
$$

## Relaciones en el álgebra $\mathcal{FK}_3$

### *Nilpotencia cuadrática*:
$$
\begin{align*}
x_{(12)}^2 = x_{(13)}^2 = x_{(23)}^2 = 0,
\end{align*}
$$
Para nosotros serán:
$$
\begin{align*}
x_{12}^2 = x_{13}^2 = x_{23}^2 = 0.
\end{align*}
$$

### Relaciones triples para $i<j<k$:

- Para $1<2<3$:
$$
x^{2}_{(ij)} \;=\; x_{(ij)}x_{(ik)} \;+\; x_{(jk)}x_{(ij)} \;+\; x_{(ik)}x_{(jk)}
\;=\;
x_{(ik)}x_{(ij)} \;+\; x_{(ij)}x_{(jk)} \;+\; x_{(jk)}x_{(ik)}
\;=\; 0,
\quad \text{(dada por } B(V)).
$$

$$
\begin{aligned}
x_{(12)}x_{(13)} + x_{(23)}x_{(12)} + x_{(13)}x_{(23)} 
\quad \text{y} \quad 
x_{(13)}x_{(12)} + x_{(12)}x_{(23)} + x_{(23)}x_{(13)}.
\end{aligned}
$$

```gap
###############################################################################
# x_(ij)^2 = 0, x_(ij)x_(jk)+x_(jk)x_(ik)+x_(ik)x_(ij)=0, etc. 
# (álgebra Fomin-Kirillov n=3,
###############################################################################

# Generadores x_(ij)
gensFK3      := [ (1,2), (1,3), (2,3) ];;
xIndex       := [1..3];;
generators   := gens{[2..4]};;
relationsFK3 := List( generators, s -> s^2);;
i            := 1;;
j            := 2;;
k            := 3;;
Xij          := generators[Position(gensFK3, (i,j))];;
Xik          := generators[Position(gensFK3, (i,k))];;
Xjk          := generators[Position(gensFK3, (j,k))];;
# primera relación: x_(ij)x_(jk) + x_(jk)x_(ik) + x_(ik)x_(ij) = 0
Add(relationsFK3, Xij*Xjk + Xjk*Xik + Xik*Xij );
# segunda relación: x_(jk)x_(ij) + x_(ik)x_(jk) + x_(ij)x_(ik) = 0
Add(relationsFK3, Xjk*Xij + Xik*Xjk + Xij*Xik );
```

## Relaciones en el álgebra $\mathcal{FK}_3$ dual

### *Nilpotencia cuadrática*:
$$
\begin{align*}
y_{(12)}^2 = y_{(13)}^2 = y_{(23)}^2 = 0.\\
\end{align*}
$$
Para nosotros serán:
$$
\begin{align*}
y_{12}^2 = y_{13}^2 = y_{23}^2 = 0.
\end{align*}
$$

### Relaciones triples para $i<j<k$:

- Para $1<2<3$:
$$
\begin{align*}
    y_{(23)}y_{(13)} + y_{(12)}y_{(23)} + y_{(13)}y_{(12)} &= 0,&&\Rightarrow& y_3y_1 + y_0y_3 + y_1y_0 &= 0, \\
    y_{(13)}x_{(23)} + y_{(23)}y_{(12)} + y_{(12)}y_{(13)} &= 0,&&\Rightarrow& y_1y_3 + y_3y_0 + y_0y_1 &= 0.
\end{align*}
$$

```gap
###############################################################################
# y_(ij)^2 = 0, y_(ij)y_(jk)+y_(jk)y_(ik)+y_(ik)y_(ij)=0, etc. 
# (álgebra Fomin-Kirillov n=3,
###############################################################################

# Generadores x_(ij)
gensFK3     := [ (1,2), (1,3), (2,3) ];;
yIndex      := [12, 13, 14];;
relsFK3dual := [];;
# Paso 1: Relaciones cuadráticas x_(ij)^2 = 0
for k in [1..Length(gensFK3 )] do
    Add(relsFK3dual, [[ [yIndex[k], yIndex[k]] ], [1]]);;
od;
# Paso 2: Relaciones cúbicas de tres índices distintos
triples := [[1,2,3]];;  # único triple distinto para n=3
for t in triples do
    i := t[1];
    j := t[2];
    k := t[3];
    ij := Position(gensFK3, (i,j));
    ik := Position(gensFK3, (i,k));
    jk := Position(gensFK3, (j,k));
    # primera relación: y_(ij)y_(jk) + y_(jk)y_(ik) + y_(ik)y_(ij) = 0
    Add(relsFK3dual,[[[yIndex[ij],yIndex[jk]],[yIndex[jk],yIndex[ik]],[yIndex[ik],yIndex[ij]]],[1,1,1]]);
    # segunda relación: y_(jk)y_(ij) + y_(ik)y_(jk) + y_(ij)y_(ik) = 0
    Add(relsFK3dual,[[[yIndex[jk],yIndex[ij]],[yIndex[ik],yIndex[jk]],[yIndex[ij],yIndex[ik]]],[1,1,1]]);
od;
```

```gap
K           := CyclicGroup(3);;
H           := CyclicGroup(2);;
AutK        := AutomorphismGroup(K);;
gensAutK    := GeneratorsOfGroup(AutK);;
T           := GroupHomomorphismByImages(H, AutK, [H.1], gensAutK);;
G           := SemidirectProduct(H, T, K);;
embK        := Embedding(G, 2);;
embH        := Embedding(G, 1);;
HinG        := List([0 .. Size(H) - 1], x -> embH(H.1^x));;
allPairsInG := Concatenation(List(K, x -> List(HinG, y -> embK(x) * y)));;
```

```gap
###############################################################################
# Álgebra de Drinfeld D(S3): relaciones δ_h g = g δ_{g^{-1} h g}
# Convención de GAP: δ_h * g = g * δ_{g * h * g^-1}
###############################################################################

# ------------------------------------------------------------
# Índices y mapeos coherentes con tu sistema GBNP
# ------------------------------------------------------------
# Representación en términos de generadores σ=(12)=4, τ=(123)=5
# (debe coincidir con tus groupIndex previos)
```

```gap
###############################################################################
# Generación automática de las relaciones δ_h g = g δ_{g^{-1} h g}
###############################################################################
ConmutationElementOfGandDeltas:=function(allPairsInG, groupIndex ,deltaIndex)
    local g, h, pos_h, pos_g, delta_h, conj, pos_conj, delta_conj, g_word, m1, m2, relsDG;
    
    relsDG := [];;
    for g in allPairsInG do
        for h in allPairsInG do
            pos_g := Position(allPairsInG, g);
            pos_h := Position(allPairsInG, h);
            if pos_g = fail or pos_h = fail then
                Error("Element not found in allPairsInG.");
            fi;
            delta_h := deltaIndex[pos_h];
            conj := g^-1 * h * g;
            pos_conj := Position(allPairsInG, conj);
            if pos_conj = fail then
                Error("No conjugate found.");
            fi;
            delta_conj := deltaIndex[pos_conj];
            g_word := groupIndex[pos_g];
            m1 := Concatenation([delta_h], g_word);    # δ_h * g
            m2 := Concatenation(g_word, [delta_conj]); # g * δ_{g^-1*h*g }
            Add(relsDG, [ [m1, m2], [1, -1] ]);
        od;
    od;
    return relsDG;
end;
```

```gap
RelationsGivenByNicholsSmashkG := function(j, gensA, xIndex, groupElems)
    local relsxBosTest, g,k, action, chi_s, posAction, leftMon, rightMon;
    relsxBosTest     := [ ];;
    for g in groupElems do
        action       := g.elm * ElementFq( j );
        posAction    := Position( gensA, action!.FieldPart );
        k            := Position( gensA, j );
        chi_s        := ChiForSemidirectProduct( g.elm );
        if posAction <> fail then
            leftMon  := Concatenation( g.index, [ xIndex[k] ] );
            rightMon := Concatenation( [ xIndex[posAction] ], g.index );
            Add(relsxBosTest, [[ leftMon, rightMon ], [1, -chi_s] ] );
        fi;
    od;
    return relsxBosTest;
end;
```

## Relaciones dadas de la definción de $\mathfrak{D}$,
$$
\begin{aligned}
y_{(ij)} x_{(ij)} + x_{(ij)} y_{(ij)} &= 1 + (ij)(\delta_{(ij)} - \delta_e), \\
y_{(ik)} x_{(ij)} + x_{(ij)} y_{(jk)} &= (ij)(\delta_{(ik)} - \delta_{(ik)(ij)}),
\end{aligned}
$$
para todas las transposiciones $(ij), (ik) \in \mathbb{S}_{3}$.

- Relaciones del tipo $y_{(ij)}x_{(ij)} + x_{(ij)}y_{(ij)} = 1 + (ij)(\delta_{(ij)} - \delta_e)$:
$$
\begin{aligned}
y_{(12)}x_{(12)} + x_{(12)}y_{(12)} &= 1 + (12)(\delta_{(12)} - \delta_e),\\
y_{(13)}x_{(13)} + x_{(13)}y_{(13)} &= 1 + (13)(\delta_{(13)} - \delta_e),\\
y_{(23)}x_{(23)} + x_{(23)}y_{(23)} &= 1 + (23)(\delta_{(23)} - \delta_e),
\end{aligned}
$$

- Relaciones del tipo $y_{(ik)}x_{(ij)} + x_{(ij)}y_{(jk)} = (ij)(\delta_{(ik)} - \delta_{(ik)(ij)})$:
$$
\begin{aligned}
y_{(13)}x_{(12)} + x_{(12)}y_{(23)} &= (12)(\delta_{(13)} - \delta_{(123)}),\\
y_{(23)}x_{(12)} + x_{(12)}y_{(13)} &= (12)(\delta_{(23)} - \delta_{(132)}),\\
y_{(12)}x_{(13)} + x_{(13)}y_{(23)} &= (13)(\delta_{(12)} - \delta_{(132)}),\\
y_{(23)}x_{(13)} + x_{(13)}y_{(12)} &= (13)(\delta_{(23)} - \delta_{(123)}),\\
y_{(12)}x_{(23)} + x_{(23)}y_{(13)} &= (23)(\delta_{(12)} - \delta_{(123)}),\\
y_{(13)}x_{(23)} + x_{(23)}y_{(12)} &= (23)(\delta_{(13)} - \delta_{(132)}).
\end{aligned}
$$


```gap
GetConmutationRulesOnFK3 := function(i, j,  gensA, xIndex, groupIndex, deltaIndex, yIndex, ElementsInSDP)
    local g_i, res, pos_x_i, pos_y_j, index_g_i, pos_res,
          trsp, mji, mij, coeff, relstrsp, index_delta, k;

    pos_x_i   := xIndex[ Position( gensA, i ) ];;
    pos_y_j   := yIndex[ Position( gensA, j ) ];;
    g_i       := ElementSDP( i, 1 );;                # elemento g en F4⋊C6
    res       := (g_i^-1) * j ;;                                # elemento en F4
    index_g_i := groupIndex[ Position( ElementsInSDP, g_i ) ];;
    pos_res   := yIndex[ Position( gensA, FieldPart(res) ) ];;
    #PrintNP([ [ [pos_res] ], [1]]);
    mji       := Concatenation( [ pos_y_j ], [ pos_x_i ] );;     # Y_j * X_i
    mij       := Concatenation( [ pos_x_i ], [ pos_res ] );;     # X_i * Y_j
    relstrsp  := [ ];;
    trsp      := TransporterOfG( ElementsInSDP, i, j);;
    coeff     := [ 1, 1 ];;
    Add( relstrsp, mji );
    Add( relstrsp, mij );
    
    if i = j then
        Add( relstrsp, [ ] );;
        Add( coeff, -1 );;
    fi;
    #PrintNP([ relstrsp, coeff ]);
    for k in [ 1..Length( trsp ) ] do
        #Print("(g_i^-1) * trsp[ k ] * g_i: ", (g_i^-1) * trsp[ k ] * g_i, "\n");
        index_delta := deltaIndex[Position( ElementsInSDP, (g_i^-1) * trsp[ k ] * g_i )];
        #PrintNP(GP2NP(gens[index_delta]));
        Add( relstrsp, Concatenation( index_g_i, [ index_delta ] ));
        Add( coeff, ChiForSemidirectProduct( trsp[ k ] ) );
    od;
    return [ relstrsp, coeff ];
end;
```

El elemento de grado máximo en $\mathbb{B}$ es

```gap
CommuteYBeforeXTest := function(prod, gensA, xIndex, groupIndex, deltaIndex, yIndex, elmsG)
    local monos, coefs, resMonos, resCoefs, i, j, k,
          pos, yj, xi, rule, newMonos, newCoefs, m, c,
          pref, suf, mon, coef, Penultimate, list_mon;
    monos       := prod[1];
    coefs       := prod[2];
    resMonos    := [];
    resCoefs    := [];
    Penultimate := list_mon -> list_mon[Length(list_mon) - 1];
    for k in [1..Length(monos)] do
        mon     := monos[k];
        coef    := coefs[k];
        xi      := Last( mon );
        if Length(mon) > 1 then
            yj  := Penultimate( mon );
            pos := Length( mon ) - 1;
            if yj in yIndex and xi in xIndex then
                j    := gensA[Position(yIndex, yj)];
                i    := gensA[xi];
                rule := GetConmutationRulesOnFK3(
                    i,
                    j,
                    gensA,
                    xIndex,
                    groupIndex,
                    deltaIndex,
                    yIndex,
                    elmsG
                );;
                newMonos := [];
                newCoefs := [];
                pref     := mon{[1..pos-1]};
                suf      := mon{[pos+2..Length(mon)]};
                for m in [3..Length(rule[1])] do
                    Add(resMonos, Concatenation(pref, rule[1][m], suf));
                    Add(resCoefs, - coef * rule[2][m]);
                od;
            else
                Add(resMonos, mon);
                Add(resCoefs, coef);
            fi;
        else
            Add(resMonos, mon);
            Add(resCoefs, coef);
        fi;
    od;
    return [resMonos, resCoefs];
end;
```

# The simple Verma modules

Let us take the expression
$$
\Omega := -2\tau^2\delta_e + 2\tau\sigma\delta_e + 4\sigma\tau\delta_e - 2\tau\delta_e + 2\sigma\delta_e - 2\sigma\tau - 2
4$$. 
Since $\delta_e$ acts as the identity in all cases, we can formally replace
$$
\delta_e \mapsto I.
$$ 
Then
$$
\Omega = -2\tau^2 + 2\tau\sigma + 4\sigma\tau - 2\tau + 2\sigma - 2\sigma\tau - 2.
$$
Grouping:
$$
\Omega = -2\tau^2 + 2\tau\sigma + 2\sigma\tau - 2\tau + 2\sigma - 2.
$$
We will use:
$$
\sigma = (12), \quad \tau = (123).
$$
In $\mathbb{S}_3$,
$$
\tau^2 = (132),
$$
and furthermore
\begin{align*}
\tau\sigma &= (123)(12) = (13), \\
\sigma\tau &= (12) (123) = (23).
\end{align*}
Therefore,
$$
\Omega = -2(132) + 2(13) + 2(23) - 2(123) + 2(12) - 2e.
$$

## The Verma modules $\mathrm{M}(e, -)$ is $\mathfrak{D}$-simple.

- 1) Sign representation on $|e\rangle_{-}$

The sign representation satisfies
$$
\operatorname{sgn}(\text{transposition}) = -1, \quad \operatorname{sgn}(\text{3-cycle}) = +1.
$$

Then:
\begin{align*}
\rho(\tau) = +1, \quad \rho(\tau^2) = +1, \\
\rho(\sigma) = -1, \quad \rho(\tau\sigma) = -1, \quad \rho(\sigma\tau) = -1.
\end{align*}
We apply this term by term:
\begin{align*}
-2\tau^2 &\mapsto -2(+1) = -2, \\
2\tau\sigma &\mapsto 2(-1) = -2, \\
2\sigma\tau &\mapsto 2(-1) = -2, \\
-2\tau &\mapsto -2(+1) = -2, \\
2\sigma &\mapsto 2(-1) = -2, \\
-2e &\mapsto -2.
\end{align*}
Adding up:
$$
-2 - 2 - 2 - 2 - 2 - 2 = -12.
$$
Therefore,
$$
\Omega|e\rangle_- = -12 |e\rangle_{-}.
$$

```gap
DeltaActionsFiltered4BasisTest:=function( prod, simple, deltaIndex, nX, nElemOfG)
    local monos, coefs, resMonos, resCoefs, conj, k, all_deltas,
    xi, mon, coef, nG, IsConstantList, conj_basis;

    monos      := prod[1];
    coefs      := prod[2];
    resMonos   := [];
    resCoefs   := [];
    all_deltas := List(GetElementsOfG(), x -> DeltaFunctionForSDP(x));
    conj_basis := Conj4Basis( simple );
    IsConstantList := function(L)
        return ForAll(L, x -> x = L[1]);
    end;
    for k in [ 1..Length(monos) ] do
        mon  := monos[k];
        coef := coefs[k];
        xi   := Last( mon );
        if xi in deltaIndex then
            if IsConstantList(conj_basis) then
                conj := conj_basis[1];
                if all_deltas[ xi - ( nX + nElemOfG ) ]( conj ) <> 0 then
                    Add( resMonos, mon );
                    Add( resCoefs, coef );
                fi;
            else
                for conj in conj_basis do
                    if all_deltas[ xi - ( nX + nElemOfG ) ]( conj ) <> 0 then
                        Add( resMonos, mon );
                        Add( resCoefs, coef );
                    fi;
                od;
            fi;
        else
            Add( resMonos, mon );
            Add( resCoefs, coef );
        fi;
    od;
    return [ resMonos, resCoefs ];
end;
```

```gap
MatrixActionYiOnNicholsBasisTest := function(prod, gensA, simple, xIndex, groupIndex, deltaIndex, yIndex, nElemOfG, BaseNichols, list_mat_in_DG)
    local Yres, Yred, nX, info, N, m, M_lambda, q, p, a, b, Yibq, I_iq, I_iq_Mat, I_iq_p, entry;
    # Paso 1: reescritura y acción
    nX   := Length(xIndex);
    Yres := RemoveAndCommuteYInYjBisTest(prod, gensA, xIndex, groupIndex, deltaIndex, yIndex, elmsG);;
    Yred := ReducedDeltaActionTest( Yres, simple, deltaIndex, nX, nElemOfG );;
    info := List(Yred, CollectLinearCombination);;
    N                     := Length(BaseNichols);;
    m                     := Length(simple.base);;
    M_lambda              := NullMat(N * m, N * m);;
    # Construcción de la matriz
    for q in [1..N] do
        Yibq     := info[q];
        I_iq     := GetInfoList(Yibq, BaseNichols, nX);
        I_iq_Mat := IiqToMatrix(I_iq, simple, list_mat_in_DG);;
        for p in [1..Length(I_iq_Mat)] do
            I_iq_p := I_iq_Mat[p];
            for a in [1..m] do
                for b in [1..m] do
                    entry := I_iq_p[a][b];
                    if entry <> 0 then
                        M_lambda[ N * (a - 1) + p ][ N * (b - 1) + q ] := entry;
                    fi;
                od;
            od;
        od;
    od;
    return M_lambda;
end;
```
