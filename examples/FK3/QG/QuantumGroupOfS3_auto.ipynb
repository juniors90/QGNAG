# El álgebra $\mathfrak{D}=\mathcal{D}(\mathcal{FK}_m\# \Bbbk\mathbb{S}_{3})$
Sea $\mathcal{FK}_m$ el álgebra presentada por generadores

$$
\{ x_{(ij)} \mid i<j \in \mathbb{I}_m \}
$$
y relaciones
$$
\begin{align*}
x_{(ij)}^2 &= 0, && i<j \in \mathbb{I}_m, \\
x_{(ij)}x_{(kl)} - x_{(kl)}x_{(ij)} &= 0, && i,j,k,l \in \mathbb{I}_m, \text{ todos distintos}, \\
x_{(jk)}x_{(ik)} - x_{(ij)}x_{(jk)} + x_{(ik)}x_{(ij)} &= 0, && i<j<k \in \mathbb{I}_m, \\
x_{(ik)}x_{(jk)} - x_{(jk)}x_{(ij)} + x_{(ij)}x_{(ik)} &= 0, && i<j<k \in \mathbb{I}_m.
\end{align*}
$$

## El álgebra $\mathcal{FK}_3$


Para $m=3$, el álgebra $\mathcal{FK}_3$ está generada por
$$
x_{(12)},\quad x_{(13)},\quad x_{(23)}
$$
y las relaciones son:
$$
\begin{aligned}
x_{(12)}^2 &= 0, \\
x_{(13)}^2 &= 0, \\
x_{(23)}^2 &= 0, \\
x_{(23)}x_{(13)} - x_{(12)}x_{(23)} + x_{(13)}x_{(12)} &= 0, \\
x_{(13)}x_{(23)} - x_{(23)}x_{(12)} + x_{(12)}x_{(13)} &= 0.
\end{aligned}
$$

## Relaciones en el álgebra $\mathcal{FK}_3$

### *Nilpotencia cuadrática*:
$$
\begin{align*}
x_{(12)}^2 = x_{(13)}^2 = x_{(23)}^2 = 0,
\end{align*}
$$
Para nosotros serán:
$$
\begin{align*}
x_{12}^2 = x_{13}^2 = x_{23}^2 = 0.
\end{align*}
$$

### Relaciones triples para $i<j<k$:

- Para $1<2<3$:
$$
x^{2}_{(ij)} \;=\; x_{(ij)}x_{(ik)} \;+\; x_{(jk)}x_{(ij)} \;+\; x_{(ik)}x_{(jk)}
\;=\;
x_{(ik)}x_{(ij)} \;+\; x_{(ij)}x_{(jk)} \;+\; x_{(jk)}x_{(ik)}
\;=\; 0,
\quad \text{(dada por } B(V)).
$$

$$
\begin{aligned}
x_{(12)}x_{(13)} + x_{(23)}x_{(12)} + x_{(13)}x_{(23)} 
\quad \text{y} \quad 
x_{(13)}x_{(12)} + x_{(12)}x_{(23)} + x_{(23)}x_{(13)}.
\end{aligned}
$$

```gap
###############################################################################
# x_(ij)^2 = 0, x_(ij)x_(jk)+x_(jk)x_(ik)+x_(ik)x_(ij)=0, etc. 
# (álgebra Fomin-Kirillov n=3,
###############################################################################

# Generadores x_(ij)
gensFK3    := [ (1,2), (1,3), (2,3) ];;
xIndex     := [1..3];;
generators := gens{[2..4]};;
relationsFK3    := List( generators, s -> s^2);;
i          := 1;;
j          := 2;;
k          := 3;;
Xij        := generators[Position(gensFK3, (i,j))];;
Xik        := generators[Position(gensFK3, (i,k))];;
Xjk        := generators[Position(gensFK3, (j,k))];;
# primera relación: x_(ij)x_(jk) + x_(jk)x_(ik) + x_(ik)x_(ij) = 0
Add(relationsFK3, Xij*Xjk + Xjk*Xik + Xik*Xij );
# segunda relación: x_(jk)x_(ij) + x_(ik)x_(jk) + x_(ij)x_(ik) = 0
Add(relationsFK3, Xjk*Xij + Xik*Xjk + Xij*Xik );
```

```gap
relsFK3 := GP2NPList(relationsFK3);;
PrintNPList( relsFK3 );
#X12^2 
#X13^2 
#X23^2 
#X12X23 + X23X13 + X13X12 
#X23X12 + X13X23 + X12X13 
```

## Relaciones en el álgebra $\mathcal{FK}_3$ dual

### *Nilpotencia cuadrática*:
$$
\begin{align*}
y_{(12)}^2 = y_{(13)}^2 = y_{(23)}^2 = 0.\\
\end{align*}
$$
Para nosotros serán:
$$
\begin{align*}
y_{12}^2 = y_{13}^2 = y_{23}^2 = 0.
\end{align*}
$$

### Relaciones triples para $i<j<k$:

- Para $1<2<3$:
$$
\begin{align*}
    y_{(23)}y_{(13)} + y_{(12)}y_{(23)} + y_{(13)}y_{(12)} &= 0,&&\Rightarrow& y_3y_1 + y_0y_3 + y_1y_0 &= 0, \\
    y_{(13)}x_{(23)} + y_{(23)}y_{(12)} + y_{(12)}y_{(13)} &= 0,&&\Rightarrow& y_1y_3 + y_3y_0 + y_0y_1 &= 0.
\end{align*}
$$

```gap
###############################################################################
# y_(ij)^2 = 0, y_(ij)y_(jk)+y_(jk)y_(ik)+y_(ik)y_(ij)=0, etc. 
# (álgebra Fomin-Kirillov n=3,
###############################################################################

# Generadores x_(ij)
gens := [ (1,2), (1,3), (2,3) ];;
yIndex := [12, 13, 14];;


# Inicializar lista de relaciones
relsFK3dual := [];;

# Paso 1: Relaciones cuadráticas x_(ij)^2 = 0
for k in [1..Length(gens)] do
    Add(relsFK3dual, [[ [yIndex[k], yIndex[k]] ], [1]]);;
od;

# Paso 2: Relaciones cúbicas de tres índices distintos
triples := [[1,2,3]];;  # único triple distinto para n=3
for t in triples do
    i := t[1]; j := t[2]; k := t[3];
    ij := Position(gens, (i,j));
    ik := Position(gens, (i,k));
    jk := Position(gens, (j,k));
    # primera relación: y_(ij)y_(jk) + y_(jk)y_(ik) + y_(ik)y_(ij) = 0
    Add(relsFK3dual,[[[yIndex[ij],yIndex[jk]],[yIndex[jk],yIndex[ik]],[yIndex[ik],yIndex[ij]]],[1,1,1]]);
    # segunda relación: y_(jk)y_(ij) + y_(ik)y_(jk) + y_(ij)y_(ik) = 0
    Add(relsFK3dual,[[[yIndex[jk],yIndex[ij]],[yIndex[ik],yIndex[jk]],[yIndex[ij],yIndex[ik]]],[1,1,1]]);
od;
```

## Relaciones dadas por $\Bbbk^{\mathbb{S}_{3}}$.

```gap
RelationsOfkGdual := function( yIndex )
    local result, i, j, entry, monPart, coeffPart, x;
    result := [];;
    for i in yIndex do
        for j in yIndex do
            if i = j then
                entry := [ [ [i, j], [i] ], [1, -1] ];
            else
                entry := [ [ [i, j] ], [1] ];
            fi;
            Add(result, entry);;
        od;
    od;
    monPart := List(yIndex, i -> [i]);;
    Add(monPart, []);
    coeffPart := Concatenation(List([1..Length(monPart)-1], x -> 1), [-1]);;
    Add(result, [ monPart, coeffPart ]);;
    return result;
end;;
```

```gap
###############################################################################
# Álgebra de Drinfeld D(S3): relaciones δ_h g = g δ_{g^{-1} h g}
# Convención de GAP: δ_h * g = g * δ_{g * h * g^-1}
###############################################################################

# ------------------------------------------------------------
# Índices y mapeos coherentes con tu sistema GBNP
# ------------------------------------------------------------
allPerms := [ (), (1,2), (1,3), (2,3), (1,2,3), (1,3,2) ];;
deltaIndex := [6..11];;

# Representación en términos de generadores σ=(12)=4, τ=(123)=5
# (debe coincidir con tus groupIndex previos)
groupIndex := [ 
    [],          # e
    [4],         # (12)
    [4,5,5],     # (13)
    [4,5],       # (23)
    [5],         # (123)
    [5,5]        # (132)
];;

relsDG := [];;

###############################################################################
# Generación automática de las relaciones δ_h g = g δ_{g^{-1} h g}
###############################################################################

for g in allPerms do
    for h in allPerms do
        pos_g := Position(allPerms, g);
        pos_h := Position(allPerms, h);
        if pos_g = fail or pos_h = fail then
            Error("Elemento no encontrado en allPerms");
        fi;

        # δ_h
        delta_h := deltaIndex[pos_h];

        # Conjugado: g * h * g^-1 (ajustado a convención GAP)
        conj := g * h * g^-1;
        pos_conj := Position(allPerms, conj);
        if pos_conj = fail then
            Error("No se encontró conjugado");
        fi;

        delta_conj := deltaIndex[pos_conj];

        # Palabra de grupo correspondiente a g
        g_word := groupIndex[pos_g];

        # monomios
        m1 := Concatenation([delta_h], g_word);        # δ_h * g
        m2 := Concatenation(g_word, [delta_conj]);     # g * δ_{g*h*g^-1}

        # Relación δ_h g - g δ_{g*h*g^-1} = 0
        Add(relsDG, [ [m1, m2], [1, -1] ]);
    od;
od;
```

## Relaciones dadas por la the bosonizaciones $\mathcal{FK}_3\#\Bbbk\mathbb{S}_{3}$

## Relaciones $gx_{(ij)} = sgn(g) x_{g(ij)g^{-1}} g$, evaluadas para todos los $g \in S_3$

### En términos de $x_{(ij)}$ para todo $g \in \mathbb{S}_3$


- Caso $(ij) = (12)$:

$$
\begin{aligned}
e\,x_{(12)} &= x_{(12)}\,e, \\[4pt]
(12)\,x_{(12)} &= -\,x_{(12)}\,(12), \\[4pt]
(123)\,x_{(12)} &= x_{(23)}\,(123), \\[4pt]
(132)\,x_{(12)} &= x_{(13)}\,(132), \\[4pt]
(23)\,x_{(12)} &= -\,x_{(13)}\,(23), \\[4pt]
(13)\,x_{(12)} &= -\,x_{(23)}\,(13).
\end{aligned}
$$

- Caso $(ij) = (13)$

$$
\begin{aligned}
e\,x_{(13)} &= x_{(13)}\,e, \\[4pt]
(12)\,x_{(13)} &= -\,x_{(23)}\,(12), \\[4pt]
(123)\,x_{(13)} &= x_{(12)}\,(123), \\[4pt]
(132)\,x_{(13)} &= x_{(23)}\,(132), \\[4pt]
(23)\,x_{(13)} &= -\,x_{(12)}\,(23), \\[4pt]
(13)\,x_{(13)} &= -\,x_{(13)}\,(13).
\end{aligned}
$$

- Caso $(ij) = (23)$:

$$
\begin{aligned}
e\,x_{(23)} &= x_{(23)}\,e, \\[4pt]
(12)\,x_{(23)} &= -\,x_{(13)}\,(12), \\[4pt]
(123)\,x_{(23)} &= x_{(13)}\,(123), \\[4pt]
(132)\,x_{(23)} &= x_{(12)}\,(132), \\[4pt]
(23)\,x_{(23)} &= -\,x_{(23)}\,(23), \\[4pt]
(13)\,x_{(23)} &= -\,x_{(12)}\,(13).
\end{aligned}
$$


```gap
###############################################################################
# Relación g * x_(ij) = sign(g) * x_(g(ij)g^-1) * g   (álgebra Fomin-Kirillov)
# con g escrito en términos de los generadores
# σ = (12), τ = (123)
###############################################################################

# Generadores x_(ij)
gens := [ (1,2), (1,3), (2,3) ];;
xIndex := [ 1, 2, 3 ];;

# Elementos del grupo
groupElems := [ (), (1,2), (1,3,2), (1,2,3), (2,3), (1,3)  ];;
groupIndex := [ [], [4],   [5, 5],     [5],  [4,5], [4,5,5] ];;  # ejemplo; adaptar si querés τ y σ


# Inicializar lista de relaciones
relsxBos := [];;

# Llenar la lista I con todas las relaciones g * x_(ij) - sign(g)*x_(g(ij)g^-1)*g = 0
for k in [1..Length(gens)] do
    h := gens[k];
    for g in groupElems do
        conj := g * h * g^-1; # Conjugación
        s := SignPerm(g);
        posConj := Position(gens, conj);
        if posConj <> fail then
            leftMon := Concatenation(groupIndex[Position(groupElems,g^-1)], [xIndex[k]]);
            rightMon := Concatenation([xIndex[posConj]], groupIndex[Position(groupElems,g^-1)]);
            Add(relsxBos, [[ leftMon, rightMon], [1, -s] ]);  # estructura válida para PrintNP
        fi;
    od;
od;
```

```gap
# Probar la primera relación
Print("Se generaron ", Length(relsxBos), " relaciones en total.\n");
PrintNPList(relsxBos);
```

```gap
# ------------------------------------------------------------
# Relaciones \delta_h xij = xij  D(ij)h}
# con h escrito en términos de los generadores
# σ = (12), τ = (123) con S_3 = { e, σ, τ, τ^2, στ, στ^2 } y sabiendo que: στ = (23), στ^2 = (13)
# ------------------------------------------------------------
###############################################################################
# Automatización de relaciones δ_h x_(ij) = x_(ij) δ_{(ij)h} para S3
# Convención con σ=(12)=4, τ=(123)=5
###############################################################################

# Generadores x_(ij)
xIndex := [1,2,3];;  # X12, X13, X23
xPerms := [ (1,2), (1,3), (2,3) ];;

# Generadores δ_h (índices en GBNP)
deltaIndex := [6..11];; # δ_e, δ_12, δ_13, δ_23, δ_123, δ_132

# Todos los elementos de S3
allPerms := [ (), (1,2), (1,3), (2,3), (1,2,3), (1,3,2) ];;

# Mapeo explícito perm -> índice δ en GBNP, según tu convención
# Orden: (), (12), (13), (23), (123), (132)
permToDelta := [7, 6, 10, 11, 9, 8];;

relsDx := [];;

###############################################################################
# Generación automática de δ_h x_(ij) = x_(ij) δ_{(ij) h}
###############################################################################

for xi in [1..3] do
    g_x := xPerms[xi];        # (ij)
    x_i := xIndex[xi];        # índice de X12,X13,X23

    for h in allPerms do
        # δ_h
        pos_h := Position(allPerms, h);
        delta_h := permToDelta[pos_h];

        # Producto (ij)*h
        prod := g_x * h;

        # δ_{(ij) h} según tu convención
        pos_prod := Position(allPerms, prod);
        delta_prod := permToDelta[pos_prod];

        # Monomios: δ_h * x_(ij) y x_(ij) * δ_{(ij)h}
        m1 := Concatenation([delta_h], [x_i]);
        m2 := Concatenation([x_i], [delta_prod]);

        # Relación δ_h x_(ij) - x_(ij) δ_{(ij)h} = 0
        Add(relsDx, [ [m1, m2], [1, -1] ]);
    od;
od;
```

\end{aligned}
$$


```gap
###############################################################################
# Relación y_(ij) * g = sign(g) * g * y_(g(ij)g^-1) (álgebra Fomin-Kirillov)
# con g escrito en términos de los generadores
# σ = (12), τ = (123)
###############################################################################

# Generadores y_(ij)
gens := [ (1,2), (1,3), (2,3) ];;
yIndex := [ 12, 13, 14 ];;

# Elementos del grupo
groupElems := [ (), (1,2), (1,2,3), (1,3,2), (2,3), (1,3)  ];;
groupIndex := [ [], [4],   [5],     [5, 5],  [4,5], [4,5,5] ];;  # ejemplo; adaptar si querés τ y σ

# Inicializar lista de relaciones
relsyBos := [];;

# Llenar la lista I con todas las relaciones g * x_(ij) - sign(g)*x_(g(ij)g^-1)*g = 0
for k in [1..Length(gens)] do
    h := gens[k];
    for g in groupElems do
        conj := g * h * g^-1; # Conjugación
        s := SignPerm(g);
        posConj := Position(gens, conj);
        if posConj <> fail then
            leftMon := Concatenation([yIndex[k]], groupIndex[Position(groupElems,g)]);
            rightMon := Concatenation( groupIndex[Position(groupElems,g)], [yIndex[posConj]]);
            Add(relsyBos, [[ leftMon, rightMon], [1, -s] ]);  # estructura válida para PrintNP
        fi;
    od;
od;
```

$$

para todas las transposiciones $(ij), (ik) \in \mathbb{S}_{3}$.

```gap
CreateTranspositions := function(n)
    return List( Combinations([1..n], 2), t -> ( t[1], t[2] ) );
end;
```

```gap
TransporterOfS3 := function( G, sigma, tau )
    local trsp, h;
    
    # Trabajamos en S_3 (puedes cambiarlo a SymmetricGroup(n) si usas más elementos)
    trsp := [];
    
    # Buscamos los h que cumplen h * sigma * h^-1 = tau (escrito al revés en GAP)
    for h in G do
        if h^-1 * sigma * h = tau then
            Add( trsp, h );
        fi;
    od;
    return trsp;
end;
```

```gap
GetConmutationRulesOnFK3 := function(ij, ik, xIndex, groupIndex, deltaIndex, yIndex)
    local gensA, g_i, res, pos_x_ij, pos_y_ik, index_g_i,
        pos_res, trsp, mji, mij, coeff,
          relstrsp, index_delta, k, allPerms;
    gensA     := CreateTranspositions( 3 );  # debe ser parametro
    pos_x_ij  := xIndex[ Position( gensA, ij ) ];;
    pos_y_ik  := yIndex[ Position( gensA, ik ) ];;
    allPerms  := [ (), (1,2), (1,3), (2,3), (1,2,3), (1,3,2) ];; # debe ser parametro
    index_g_i := groupIndex[ Position( allPerms, ij ) ];;
    g_i       := ij;
    res       := g_i * ik * g_i;; # g_i actuando sobre ik por conjugacion 
    pos_res   := yIndex[ Position( gensA, res ) ];
    mji       := Concatenation( [ pos_y_ik ], [ pos_x_ij ] );; # Y_j * X_i
    mij       := Concatenation( [ pos_x_ij ], [ pos_res ] );;  # X_i * Y_j
    trsp      := TransporterOfS3( allPerms, ij, ik ); # TransporterOfS3 debe ser param TransporterOfG
    relstrsp  := [ ];;
    coeff     := [ 1, 1 ];;
    Add( relstrsp, mji );
    Add( relstrsp, mij );
    if ij = ik then
        Add( relstrsp, [ ] );;
        Add( coeff, -1 );;
    fi; 
    for k in [ 1..Length( trsp ) ] do
        index_delta := deltaIndex[Position( allPerms, (g_i^-1) * trsp[ k ] * g_i )];
        Add( relstrsp, Concatenation( index_g_i, [ index_delta ] ));
        Add( coeff, SignPerm( trsp[ k ] ) ); # SignPerm debe ser param chi
    od;
    return [ relstrsp, coeff ];
end;
```

## Calculo de $y_{top}x_{top}$

Las relaciones definitorias del álgebra de Nichols $\mathfrak{B}(V)$ son
$$
x_{(12)}^{2}, \quad x_{(13)}^{2}, \quad x_{(23)}^{2},
$$
$$
x_{(12)}x_{(13)} + x_{(23)}x_{(12)} + x_{(13)}x_{(23)} 
\quad \text{y} \quad 
x_{(13)}x_{(12)} + x_{(12)}x_{(23)} + x_{(23)}x_{(13)}.
$$

Denotamos por $\mathbb{B}$ a la base de $\mathfrak{B}(V)$, la cual se obtiene eligiendo un elemento por cada fila de la siguiente lista y multiplicándolos de arriba hacia abajo, véase *e.g.* [7]:

\begin{array}{l}
1, \, x_{(12)}, \\[4pt]
1, \, x_{(13)}, \, x_{(13)}x_{(12)}, \\[4pt]

\end{array}
$$

Sea $\mathbb{B}^n = \mathbb{B} \cap \mathfrak{B}^n(V)$, $n \geq 0$. 
El elemento de grado máximo en $\mathbb{B}$ es
$$
x_{\text{top}} = x_{(12)}x_{(13)}x_{(12)}x_{(23)} \in \mathbb{B}^4.
$$

```gap
CommuteYBeforeX:=function(prod)
    local monos, coefs, resMonos, resCoefs, xIndex, yIndex,deltaIndex, groupIndex,
        i, j, k, pos, yj, xi, rule, newMonos, gensA,
        newCoefs, m, c, pref, suf, mon, coef, Penultimate, list_mon,_monos, _coefs;
 
    monos       := prod[1];
    coefs       := prod[2];
    resMonos    := [];
    resCoefs    := [];
    Penultimate := list_mon -> list_mon[Length(list_mon) - 1];
    xIndex     := [1..3];;                               # X12, X13, X23
    groupIndex := [ [], [4], [4,5,5], [4,5], [5], [5,5] ];;
    deltaIndex := [6, 7, 8, 9, 10, 11];;                    # δ_e, δ_12, δ_13, δ_23, δ_123, δ_132 
    yIndex     := [12..14];;                            # Y12, Y13, Y23
    gensA     := CreateTranspositions( 3 );  # debe ser parametro
    for k in [1..Length(monos)] do
        mon  := monos[k];
        coef := coefs[k];
        xi   := Last( mon );
        if Length(mon) > 1 then
            # PrintNP([[mon], [coef]]);
            yj := Penultimate( mon );
            pos := Length( mon ) - 1;
            # Verificar si es un par Y_i X_j
            if yj in yIndex and xi in xIndex then
                # Obtener índices i y j reales
                j := Position(yIndex, yj);   # Y_j
                i := xi;
                                            # X_i
                # Obtener la regla de conmutación correspondiente
                rule := GetConmutationRulesOnFK3(gensA[i], gensA[j], xIndex, groupIndex, deltaIndex, yIndex);
                newMonos := [];
                newCoefs := [];
                # prefijo y sufijo del monomio alrededor de Y_j X_i
                pref := mon{[1..pos-1]};
                suf  := mon{[pos+2..Length(mon)]};
                # Expandir según la regla
                for m in [3..Length(rule[1])] do
                    #Print(rule[1][m], "\n");
                    Add(resMonos, Concatenation(pref, rule[1][m], suf));
                    Add(resCoefs, - coef * rule[2][m]);
                od;
            else
                Add(resMonos, mon);
                Add(resCoefs, coef);
            fi;
        else
            Add(resMonos, mon);
            Add(resCoefs, coef);
        fi;
    od;
    return [resMonos, resCoefs];
end;
```

```gap
RemoveYterms:=function(prod)
    local monos, coefs, resMonos, resCoefs, i, ult, yIndex;
    monos    := prod[1];
    coefs    := prod[2];
    resMonos := [];
    resCoefs := [];
    yIndex   := [12..14];; 
    for i in [1..Length(monos)] do
        ult := Last(monos[i]);  # último índice del monomio
        # si el último índice NO es Y0,Y1,Y2,Y3 (33..36), lo conservamos
        if not (ult in yIndex) then
            Add(resMonos, monos[i]);
            Add(resCoefs, coefs[i]);
        fi;
    od;
    return [resMonos, resCoefs];
end;
```

```gap
FilterMonomials := function(poly, elemIndex)
    local monomials, coeffs, newmons, newcoeffs, i, mon, hasY;

    monomials := poly[1];
    coeffs    := poly[2];

    newmons   := [];
    newcoeffs := [];

    for i in [1 .. Length(monomials)] do
        mon := monomials[i];

        # Verifica si aparece alguna Y
        hasY := ForAny(mon, x -> x in elemIndex);

        # Si NO tiene Y's, lo conservamos
        if not hasY then
            Add(newmons, mon);
            Add(newcoeffs, coeffs[i]);
        fi;
    od;

    return [ newmons, newcoeffs ];
end;
```
