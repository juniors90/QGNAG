```gap
TensorProductOfSimples := function( simple1, simple2 )
    local gens, mgens, gen1, gen2, newrho, rep;
    gens  := simple1.generatorsofgroup;
    gen1  := simple1.genimages;
    gen2  := simple2.genimages;
    mgens := List( [1..Length( gens )],
                    i -> KroneckerProduct( gen1[i], gen2[i] ) );
    newrho   := GroupHomomorphismByImages( simple1.group , Group( mgens ), gens, mgens );
    rep   := rec(
                group             := simple1.group,
                generatorsofgroup := gens,        # new feature
                rho               := newrho,      # new feature
                # irreps            := simple1.irreps, # new feature
                genimages         := mgens,
                isRepresentation  := true,
                # isIrreps          := newrho in simple1.irreps,
                dimension         := Length(mgens[1])
    );
    return rep;
end;
```

## Yetter-Drinfeld category over finite groups.
Let $G$ be a finite group.
A Yetter-Drinfeld module over the group algebra $\Bbbk G$ is a $G$-graded vector space
$$
V=\bigoplus_{g\in G}V_g
$$
which is also a left $\Bbbk G$-module such that
$$
h\cdot V_g \subseteq V_{hgh^{-1}}
$$
for all $g,h\in G$. Denote the category of Yetter-Drinfeld modules over $\Bbbk G$ by
$$
{}_G^G\mathcal{YD}.
$$
The category ${}_G^G\mathcal{YD}$ is braided. For $V,W\in {}_G^G\mathcal{YD}$, the braiding
$$
c_{V,W}:V\otimes W\to W\otimes V
$$
is given by
$$
c_{V,W}(v\otimes w)=g\cdot w\otimes v,
\qquad v\in V_g,\; w\in W.
$$

## Yetter-Drinfeld category over finite groups.
Let $G$ be a finite group. A Yetter-Drinfeld module over the group algebra $\Bbbk G$ is a left $\Bbbk G$-module $V$ together with a left $\Bbbk G$-coaction
$$
\rho:V\to \Bbbk G\otimes V
$$
such that
$$
\rho(h\cdot v)=h v_{(-1)} h^{-1}\otimes h\cdot v_{(0)}
$$
for all $h\in G$ and $v\in V$, where
$$
\rho(v)=v_{(-1)}\otimes v_{(0)}.
$$
Equivalently,
$$
V=\bigoplus_{g\in G}V_g,
\qquad
V_g=\{v\in V\mid \rho(v)=g\otimes v\},
$$
and the compatibility condition becomes
$$
h\cdot V_g\subseteq V_{hgh^{-1}}.
$$

Denote this category by ${}_G^G\mathcal{YD}$.
