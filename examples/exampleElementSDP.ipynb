```gap
QGNAG_ActionTable := function( Gels, Xels )
    local tbl, row, g, x, widths, i, j, strtbl, s;
    # convertir todo a strings
    strtbl := [];
    # encabezado
    row := Concatenation( [ "( a, g ) \\ x" ], List( Xels, x -> String(x) ) );
    Add( strtbl, row );
    # filas
    for g in Gels do
        row := [ String(g) ];
        for x in Xels do
            Add( row, String( g * x ) );
        od;
        Add( strtbl, row );
    od;
    # anchos máximos por columna
    widths := [];
    for j in [1..Length(strtbl[1])] do
        widths[j] := Maximum( List( strtbl, r -> Length(r[j]) ) );
    od;
    # imprimir tabla alineada
    for row in strtbl do
        for j in [1..Length(row)] do
            s := row[j];
            Print(s, String( ListWithIdenticalEntries(widths[j] - Length(s) + 3, ' ' ) ) );
        od;
        Print("\n");
    od;
    return;
end;
```

```gap
LatexActionTable:=function( Gels, Xels )

    local tex, row, g, x, val, StrElem;

    # conversión linda a latex
    StrElem := function(z)

        local s;

        s := String(z);
        s := ReplacedString(s, "0*Z(2)", "0");
        s := ReplacedString(s, "Z(2)^0", "1");
        s := ReplacedString(s, "Z(2^2)", "\\omega");
        s := ReplacedString(s, "Z(2^2)^2", "\\omega^2");

        return s;

    end;

    tex := "";

    tex := Concatenation(
        tex,
        "\\[\n",
        "\\begin{array}{c|",
        Concatenation( List([1..Length(Xels)], i -> "c") ),
        "}\n"
    );

    # encabezado
    tex := Concatenation( tex, "(a,g)" );

    for x in Xels do
        tex := Concatenation(
            tex,
            " & ",
            StrElem(x)
        );
    od;

    tex := Concatenation( tex, "\\\\\n\\hline\n" );

    # filas
    for g in Gels do

        tex := Concatenation(
            tex,
            String(g)
        );

        for x in Xels do

            val := g * x;

            tex := Concatenation(
                tex,
                " & ",
                StrElem(val)
            );

        od;

        tex := Concatenation(tex, "\\\\\n");

    od;

    tex := Concatenation(
        tex,
        "\\end{array}\n",
        "\\]\n"
    );

    Print(tex);

end;
```

```gap
QGNAG_ActionTable := function( Gels, Xels )

    local row, g, x, widths, j, strtbl, s, pad;

    # tabla de strings
    strtbl := [];
    # encabezado
    row := Concatenation( [ "( a , g ) \\ x " ], List( Xels, x -> String(ElementFq(x)) ) );
    Add( strtbl, row );
    # filas
    for g in Gels do
        row := [ String(g) ];
        for x in Xels do
            Add( row, String( g * x ) );
        od;
        Add( strtbl, row );
    od;
    # anchos máximos
    widths := [];
    for j in [1..Length(strtbl[1])] do
        widths[j] := Maximum( List( strtbl, r -> Length(r[j]) ) );
    od;
    # imprimir
    for row in strtbl do
        for j in [1..Length(row)] do
            s := row[j];
            Print(s);
            # padding manual
            pad := widths[j] - Length(s) + 2;
            while pad > 0 do
                Print(" ");
                pad := pad - 1;
            od;
        od;
        Print("\n");
    od;
    return;
end;
```
