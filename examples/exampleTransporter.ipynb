```gap
GetElementsOfG:=function()
    local l, k, p, elmsofG;
    elmsofG:=[];;
    for l in Elements(GF(ConfigSDP.p)) do
        for k in [0..(ConfigSDP.m-1)] do
            Add(elmsofG, ElementSDP(l, k));
        od;
    od;
    return elmsofG;
end;
```

```gap
Alias(GetElementsOfG()[3]);
```

```gap
trsp := TransporterOfG( GetElementsOfG(), r, s);
```
